## الخطوة ٥ — السؤال الرابع | Step 5 — Question Four

### 🔍 السؤال ٤: احسب "نسبة الفجوة" لكل لغة — نمو الإنترنت مقسوماً على تمثيل الذكاء الاصطناعي
### 🔍 Question 4: Calculate the "gap ratio" for each language — internet growth divided by AI representation

كلما كانت النسبة أعلى، كلما كانت الفجوة أكبر بين حجم المجتمع وتمثيله في الذكاء الاصطناعي.
The higher the ratio, the bigger the gap between community size and AI representation.

````python
# احسب نسبة الفجوة | Calculate the gap ratio
df['gap_ratio'] = df['growth_pct'] / df['ai_training_data_pct']

# رتّب حسب الفجوة الأكبر | Sort by largest gap
gap_ranking = df.sort_values('gap_ratio', ascending=False)

print('ترتيب الفجوة | Gap Ranking:')
print(gap_ranking[['language', 'growth_pct', 'ai_training_data_pct', 'gap_ratio']].to_string(index=False))
````

---
## الخطوة ٦ — الرسم البياني | Step 6 — The Chart

### 📊 السؤال الخامس: ارسم مخططاً يُظهر الفجوة بين نمو الإنترنت وتمثيل الذكاء الاصطناعي لأعلى ٥ لغات نمواً
### 📊 Question 5: Draw a chart showing the gap between internet growth and AI representation for the top 5 growing languages

هذا الرسم البياني هو القصة التي ستنشرها على GitHub.
This chart is the story you'll publish on GitHub.

````python
# أعلى ٥ لغات نمواً | Top 5 growing languages
top5 = df.sort_values('growth_pct', ascending=False).head(5)

# إعداد الرسم | Set up the chart
fig, ax = plt.subplots(figsize=(10, 6))

x = range(len(top5))
width = 0.35

# رسم عمودين | Draw two bars
bars1 = ax.bar([i - width/2 for i in x], 
               top5['growth_pct'] / 100,  # نقسم على 100 لتحسين الشكل
               width, 
               label='نمو الإنترنت ٪ / Internet Growth %', 
               color='#1B4FD8',
               alpha=0.8)

bars2 = ax.bar([i + width/2 for i in x], 
               top5['ai_training_data_pct'],
               width, 
               label='بيانات الذكاء الاصطناعي ٪ / AI Training Data %', 
               color='#0E7C5B',
               alpha=0.8)

# تنسيق الرسم | Format the chart
ax.set_xticks(x)
ax.set_xticklabels(top5['language'], fontsize=12)
ax.set_title('الفجوة بين نمو الإنترنت وتمثيل الذكاء الاصطناعي\nThe Gap Between Internet Growth and AI Representation', 
             fontsize=14, pad=20)
ax.legend(fontsize=10)
ax.set_ylabel('النسبة المئوية / Percentage', fontsize=11)

plt.tight_layout()

# حفظ الرسم | Save the chart
plt.savefig('my_finding.png', dpi=150, bbox_inches='tight')
plt.show()

print('✓ تم حفظ الرسم البياني | Chart saved as my_finding.png')
````
